```scala
spark.version
import com.johnsnowlabs.nlp.util.io.OcrHelper
import com.johnsnowlabs.nlp.util.io._
import com.johnsnowlabs.nlp.annotators.Tokenizer
import com.johnsnowlabs.nlp.base.{DocumentAssembler, Finisher}
```

```scala
import com.johnsnowlabs.nlp.util.io.{PageSegmentationMode, PageIteratorLevel}

OcrHelper.setScalingFactor(2.5f)
OcrHelper.setPageIteratorLevel(PageIteratorLevel.PARAGRAPH)
OcrHelper.useErosion(true, kSize = 2)
val data = OcrHelper.createDataset(spark, inputPath="../../../strata_demo",
                                   outputCol="region", metadataCol="metadata").cache
data.printSchema

{ // for displaying
val regions = data.select("region").collect().map(_.get(0))
regions.foreach{chunk =>
    println("---------------")
    println(chunk)}
}
```

```scala
import com.johnsnowlabs.nlp.annotators.spell.norvig.NorvigSweetingModel
import org.apache.spark.ml.{Pipeline, PipelineModel}


val documentAssembler = new DocumentAssembler().
setInputCol("region").
setMetadataCol("metadata").
setTrimAndClearNewLines(false)

val token = new Tokenizer().
setTargetPattern("\\S+|\\S+:|\\n|,|\\.|\\n\\n").
setIncludeDefaults(false).
setInfixPatterns(Array("(\\S+)(\\n)(\\n)(\\S+)","(\\S+)(\\n)(\\S+)", "(\\S+)(\\n)", "(\\S+)(\\.)", "(\\S+)(\\,)", "(\\S+)")).
setInputCols(Array("document")).
setOutputCol("token")


val pipeline = new Pipeline().setStages(Array(documentAssembler, token)) 
val result = pipeline.fit(data).transform(data)

result.select("token").collect.foreach(println)
```

```scala
val spellChecker = NorvigSweetingModel.load("schecker_norvig").
setInputCols("token").
setOutputCol("checked_token")

val finisher = new Finisher().
setInputCols("checked_token")

val pipeline = new Pipeline().setStages(Array(documentAssembler, token, spellChecker, finisher)) 
val result = pipeline.fit(data).transform(data)

result.select("finished_checked_token")

.as[Array[String]].collect().foreach { a => 
    println("---------------")
    println(a.mkString(" "))
}
```
